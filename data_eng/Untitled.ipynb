```python
def get_files(filepath):
    all_files=[]
    for root,dirs,files in os.walk(filepath):
        files=glob.glob(os.path.join(root,'*.json'))
        for f in files:
            all_files.append(os.path.abspath(f))
    return all_files
```

```python
time_dict={}
if len(date_set)==len(column_lables):
    for i,column in enumerate(column_lables):
        time_dict[column]=date_set[i]
else:
    if(len(date_set)>len(column_lables)):
        print('not enough lables for data')
    else:
        print('More lables specified, less data present')
```

```python
time_df=pd.DataFrame.from_dict(time_dict)
```
